```python
import torch
import cv2
import json
import os
from pathlib import Path
import matplotlib.pyplot as plt
import argparse
import numpy as np
```

```python
def test_yolov8_combined(weights, data, img_size=640, conf_thres=0.25, iou_thres=0.45, save_results=True, device=None):
    # ë””ë°”ì´ìŠ¤ ì„¤ì • (CUDA or CPU)
    device = device or ('cuda' if torch.cuda.is_available() else 'cpu')

    # ëª¨ë¸ ë¡œë“œ
    print(f"ğŸ”¹ ëª¨ë¸ ë¡œë“œ ì¤‘... ({weights})")
    model = YOLO(weights)  # YOLOv8 ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
    model.to(device)  # ë””ë°”ì´ìŠ¤ë¡œ ì´ë™

    # í…ŒìŠ¤íŠ¸í•  ì´ë¯¸ì§€ ë¦¬ìŠ¤íŠ¸
    test_images = sorted(list(Path(data).rglob("*.jpg")) +
                         list(Path(data).rglob("*.png")) +
                         list(Path(data).rglob("*.jpeg")))

    if not test_images:
        print(f"âŒ í…ŒìŠ¤íŠ¸í•  ì´ë¯¸ì§€ê°€ {data} ê²½ë¡œì— ì—†ìŠµë‹ˆë‹¤!")
        return

    print(f"ğŸ”¹ {len(test_images)}ê°œ ì´ë¯¸ì§€ í…ŒìŠ¤íŠ¸ ì‹œì‘...")

    save_dir = Path("/content/test_results")  # ê²°ê³¼ ì €ì¥ ë””ë ‰í† ë¦¬ (Colab í™˜ê²½)
    save_dir.mkdir(parents=True, exist_ok=True)

    result_images = []  # ê²°ê³¼ ì´ë¯¸ì§€ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸

    # ì´ë¯¸ì§€ë³„ ì¶”ë¡  ë° ê²°ê³¼ ì €ì¥
    for img_path in test_images:
        img = cv2.imread(str(img_path))
        if img is None:
            print(f"âŒ ì´ë¯¸ì§€ ë¡œë“œ ì‹¤íŒ¨: {img_path}")
            continue

        # ëª¨ë¸ ì¶”ë¡ 
        results = model(img, imgsz=img_size, conf=conf_thres, iou=iou_thres, device=device)

        # ê²°ê³¼ ì´ë¯¸ì§€ ìƒì„±
        result_img = results[0].plot()  # ë°”ìš´ë”© ë°•ìŠ¤ê°€ í¬í•¨ëœ ì´ë¯¸ì§€ ìƒì„±

        result_images.append(result_img)

    # ğŸ”¹ ì´ë¯¸ì§€ë¥¼ í•œ ì¤„ì— 4ê°œì”© í•©ì¹˜ê¸°
    if result_images:
        # í•œ ì¤„ì— 4ê°œì”© ë°°ì¹˜
        num_images = len(result_images)
        num_rows = (num_images + 3) // 4  # í–‰ì˜ ê°œìˆ˜ ê³„ì‚°
        max_width = max(img.shape[1] for img in result_images)
        row_height = max(img.shape[0] for img in result_images)

        # ë¹ˆ ìº”ë²„ìŠ¤ ìƒì„± (ë°°ê²½: ê²€ì •ìƒ‰)
        combined_img = np.zeros((row_height * num_rows, max_width * 4, 3), dtype=np.uint8)

        # ì´ë¯¸ì§€ë“¤ì„ í–‰ë ¬ í˜•íƒœë¡œ ë°°ì¹˜
        y_offset = 0
        for i in range(num_rows):
            x_offset = 0
            for j in range(4):
                idx = i * 4 + j
                if idx < num_images:
                    img = result_images[idx]
                    combined_img[y_offset:y_offset+row_height, x_offset:x_offset+max_width] = img
                x_offset += max_width
            y_offset += row_height

        # ìµœì¢… ê²°ê³¼ ì €ì¥
        final_result_path = save_dir / "test_results_combined.jpg"
        cv2.imwrite(str(final_result_path), combined_img)

        # Colabì—ì„œ ì´ë¯¸ì§€ í‘œì‹œ
        plt.figure(figsize=(15, num_rows * 4))
        plt.imshow(cv2.cvtColor(combined_img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

        print(f"âœ… ìµœì¢… ê²°ê³¼ ì €ì¥ ì™„ë£Œ: {final_result_path}")

    else:
        print("âŒ ê²°ê³¼ë¥¼ ì €ì¥í•  ì´ë¯¸ì§€ê°€ ì—†ìŠµë‹ˆë‹¤.")
```

```python
weights_path = "/content/runs/train/train/weights/best.pt"  # ê°€ì¤‘ì¹˜ íŒŒì¼ (Google Driveì—ì„œ ê°€ì ¸ì˜¤ê¸°)
test_images_path = "/content/drive/MyDrive/TransTRACK/DATA/images/test"  # í…ŒìŠ¤íŠ¸í•  ì´ë¯¸ì§€ í´ë”

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
test_yolov8_combined(weights=weights_path, data=test_images_path, save_results=True)
```
