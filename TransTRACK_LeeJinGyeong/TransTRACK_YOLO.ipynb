Testing the YOLO

```python
import torch
import cv2
import json
import os
from pathlib import Path
import matplotlib.pyplot as plt
import argparse
import numpy as np
```

```python
def test_yolov8_combined(weights, data, img_size=640, conf_thres=0.25, iou_thres=0.45, save_results=True, device=None):
    # 디바이스 설정 (CUDA or CPU)
    device = device or ('cuda' if torch.cuda.is_available() else 'cpu')

    # 모델 로드
    print(f"🔹 모델 로드 중... ({weights})")
    model = YOLO(weights)  # YOLOv8 모델 불러오기
    model.to(device)  # 디바이스로 이동

    # 테스트할 이미지 리스트
    test_images = sorted(list(Path(data).rglob("*.jpg")) +
                         list(Path(data).rglob("*.png")) +
                         list(Path(data).rglob("*.jpeg")))

    if not test_images:
        print(f"❌ 테스트할 이미지가 {data} 경로에 없습니다!")
        return

    print(f"🔹 {len(test_images)}개 이미지 테스트 시작...")

    save_dir = Path("/content/test_results")  # 결과 저장 디렉토리 (Colab 환경)
    save_dir.mkdir(parents=True, exist_ok=True)

    result_images = []  # 결과 이미지를 저장할 리스트

    # 이미지별 추론 및 결과 저장
    for img_path in test_images:
        img = cv2.imread(str(img_path))
        if img is None:
            print(f"❌ 이미지 로드 실패: {img_path}")
            continue

        # 모델 추론
        results = model(img, imgsz=img_size, conf=conf_thres, iou=iou_thres, device=device)

        # 결과 이미지 생성
        result_img = results[0].plot()  # 바운딩 박스가 포함된 이미지 생성

        result_images.append(result_img)

    # 🔹 이미지를 한 줄에 4개씩 합치기
    if result_images:
        # 한 줄에 4개씩 배치
        num_images = len(result_images)
        num_rows = (num_images + 3) // 4  # 행의 개수 계산
        max_width = max(img.shape[1] for img in result_images)
        row_height = max(img.shape[0] for img in result_images)

        # 빈 캔버스 생성 (배경: 검정색)
        combined_img = np.zeros((row_height * num_rows, max_width * 4, 3), dtype=np.uint8)

        # 이미지들을 행렬 형태로 배치
        y_offset = 0
        for i in range(num_rows):
            x_offset = 0
            for j in range(4):
                idx = i * 4 + j
                if idx < num_images:
                    img = result_images[idx]
                    combined_img[y_offset:y_offset+row_height, x_offset:x_offset+max_width] = img
                x_offset += max_width
            y_offset += row_height

        # 최종 결과 저장
        final_result_path = save_dir / "test_results_combined.jpg"
        cv2.imwrite(str(final_result_path), combined_img)

        # Colab에서 이미지 표시
        plt.figure(figsize=(15, num_rows * 4))
        plt.imshow(cv2.cvtColor(combined_img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

        print(f"✅ 최종 결과 저장 완료: {final_result_path}")

    else:
        print("❌ 결과를 저장할 이미지가 없습니다.")
```

```python
weights_path = "/content/runs/train/train/weights/best.pt"  # 가중치 파일 (Google Drive에서 가져오기)
test_images_path = "/content/drive/MyDrive/TransTRACK/DATA/images/test"  # 테스트할 이미지 폴더

# 테스트 실행
test_yolov8_combined(weights=weights_path, data=test_images_path, save_results=True)
```
